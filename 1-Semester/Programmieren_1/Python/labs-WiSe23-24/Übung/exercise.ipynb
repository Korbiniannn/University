```python
def merge(lst1, lst2):
    """Merges two sorted lists, lst1 and lst2. The new list contains all
    elements in sorted order.
    >>> merge([1, 3, 5], [2, 4, 6])
    [1, 2, 3, 4, 5, 6]
    >>> merge([], [2, 4, 6])
    [2, 4, 6]
    >>> merge([1, 2, 3], [])
    [1, 2, 3]
    >>> merge([5, 7], [2, 4, 6])
    [2, 4, 5, 6, 7]
    """
    # YOUR CODE HERE
    lst1.extend(lst2)
    i = 0
    while i < len(lst1):
        key = i
        while key > 0:
            if lst1[key -1] > lst1[key]:
                lst1[key -1], lst1[key] = lst1[key], lst1[key -1]
            key -= 1
        i += 1
    return lst1

import doctest
doctest.testmod()
```

```python
def caesar_cipher(message, shift):
    """
    A Caesar Cipher is a simple cipher that works by shifting
    each letter in the given message by a certain number. For
    example, if we shift the message "We Attack At Dawn" by 1
    letter, it becomes "Xf Buubdl Bu Ebxo".

    ceasar_cipher(message, shift) shifts the given message by
    'shift' many letters. You are guaranteed that 'message' is a
    string, and that 'shift' is an integer between -25 and 25.
    Capital letters should stay capital and lowercase letters
    should stay lowercase, and non-letter characters should
    not be changed."Z" wraps around to "A".
    >>> caesar_cipher("We Attack At Dawn", 1)
    'Xf Buubdl Bu Ebxo'
    >>> caesar_cipher("zodiac",-2)
    'xmbgya'
    """
    new = ""
    upper_case = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    lower_case = upper_case.lower()
    for elem in message:
        if elem == " ":
            new += elem
        if elem in upper_case:
            x = upper_case.index(elem)
            new += upper_case[x + shift]
        if elem in lower_case:
            y = lower_case.index(elem)
            new += lower_case[y + shift]
    return new
import doctest
doctest.testmod()



# YOUR CODE HERE
```

```python
def most_frequent(L):
    """
    Return the most frequent element in a list L, resolving
    ties arbitrarily. L only has elements of scalar types.
    >>> most_frequent([2,5,3,4,6,4,2,4,5])
    4
    >>> most_frequent([2,3,4,3,5,3,6,3,7])
    3
    >>> most_frequent([42])
    42
    >>> most_frequent([])
    
    """
    # YOUR CODE HERE
    if len(L) == 0:
        return None
    dct = {}
    for elem in L:
        if elem in dct.keys():
            dct[elem] += 1
        else:
            dct[elem] = 1
    current_k = 0
    current_v = 0
    for key, val in dct.items():
        if val > current_v:
            current_v = val
            current_k = key
    return current_k
import doctest
doctest.testmod()
```

```python
def count_occurrences(t, n, x):
    """Return the number of times that x is equal to one of the
    first n elements of iterator t.
    >>> s = iter([10, 9, 10, 9, 9, 10, 8, 8, 8, 7])
    >>> count_occurrences(s, 10, 9)
    3
    >>> s2 = iter([10, 9, 10, 9, 9, 10, 8, 8, 8, 7])
    >>> count_occurrences(s2, 3, 10)
    2
    >>> s = iter([3, 2, 2, 2, 1, 2, 1, 4, 4, 5, 5, 5])
    >>> count_occurrences(s, 1, 3) # Only iterate over 3
    1
    >>> count_occurrences(s, 3, 2) # Only iterate over 2, 2, 2
    3
    >>> list(s) # Ensure that the iterator has advanced the right amount
    [1, 2, 1, 4, 4, 5, 5, 5]
    >>> s2 = iter([4, 1, 6, 6, 7, 7, 6, 6, 2, 2, 2, 5])
    >>> count_occurrences(s2, 6, 6)
    2
    """
    t = iter(t)
    count = 0
    while n > 0:
        n -= 1
        if x == next(t):
            count += 1
    return count
import doctest
doctest.testmod()
        
```

```python
def repeated(t, k):
    """Return the first value in iterator t that appears k times in a row,
    calling next on t as few times as possible.
    >>> s = iter([10, 9, 10, 9, 9, 10, 8, 8, 8, 7])
    >>> repeated(s, 2)
    9
    >>> s2 = iter([10, 9, 10, 9, 9, 10, 8, 8, 8, 7])
    >>> repeated(s2, 3)
    8
    >>> s = iter([3, 2, 2, 2, 1, 2, 1, 4, 4, 5, 5, 5])
    >>> repeated(s, 3)
    2
    >>> repeated(s, 3)
    5
    >>> s2 = iter([4, 1, 6, 6, 7, 7, 8, 8, 2, 2, 2, 5])
    >>> repeated(s2, 3)
    2
    """
    assert k > 1
    t = iter(t)
    count = 1
    x = next(t)
    while count != k:
            y = next(t)
            if x == y:
                count += 1
            else:
                count = 1
            x = y
    return x

import doctest
doctest.testmod()
```

```python
def matches(a, b):
    """
    Return the number of values k such that A[k] == B[k].
    >>> matches([1, 2, 3, 4, 5], [3, 2, 3, 0, 5])
    3
    >>> matches("abdomens","indolence")
    4
    >>> matches("abcd","dcba")
    0
    >>> matches("abcde","edcba")
    1
    >>> matches("abcdefg","edcba")
    1
    """
    if len(a) < len(b):
        l = len(a)
    else:
        l = len(b)

    c = zip(a,b)
    c = iter(c)
    count = 0
    result = 0
    while count < l:
        x = next(c)
        if x[0] == x[1]:
            result += 1
        count += 1
    return result
import doctest
doctest.testmod()
```

```python
def palindrome(s):
    """Return whether s is the same sequence backward and forward.
    >>> palindrome([3, 1, 4, 1, 5])
    False
    >>> palindrome([3, 1, 4, 1, 3])
    True
    >>> palindrome('seveneves')
    True
    >>> palindrome('seven eves')
    False
    """
    l = len(s)
    s1 = s[::-1]
    lol = zip(s, s1)
    n = 0
    b = next(lol)
    while n < l-1:
        n += 1
        if b[0] == b[1]:
            c = True
        else:
            return False
        b = next(lol)
    return c
import doctest
doctest.testmod()
```

```python
def binary_search(L, x):
    middle = len(L) // 2
    if len(L) == 1:
        if L[0] == x:
            return True
        return False
    else:
        if x > L[middle]:
            return binary_search(L[middle:], x)
        if x < L[middle]:
            return binary_search(L[:middle], x)
        if x == L[middle]:
            return True
binary_search([1, 5, 4, 7, 6, 8], 10)
```

```python
def filter_iter(iterable, f):
    """
    Generates elements of iterable for which f returns True.
    >>> is_even = lambda x: x % 2 == 0
    >>> list(filter_iter(range(5), is_even))
    [0, 2, 4]
    >>> all_odd = (2*y-1 for y in range(5))
    >>> list(filter_iter(all_odd, is_even))
    []
    >>> naturals = (n for n in range(1, 100))
    >>> s = filter_iter(naturals, is_even)
    >>> next(s)
    2
    >>> next(s)
    4
    """
    for elem in iterable:
            if f(elem) == True:
                yield elem

import doctest
doctest.testmod()
```

```python
def setKthDigit(n, k, d):
    """
    n is an integer, k is a non-negative integer and d is non-negative
    single digit (0 = d = 9). Return the number n with the kth digit
    replaced with d.
    >>> setKthDigit(468, 0, 1)
    461
    >>> setKthDigit(468, 1, 1)
    418
    >>> setKthDigit(468, 2, 1)
    168
    >>> setKthDigit(468, 3, 1)
    1468
    """
    n = str(n)
    rev_n = n[::-1]
    if k in range(len(n)):
        result = n.replace(rev_n[k], str(d))
        return int(result)
    else:
        result = (rev_n + str(d))[::-1]
        return int(result)

import doctest 
doctest.testmod()
```
