```python
import math
import numpy as np
import matplotlib.pyplot as plt

# Define the function f
def f(x):
    return math.exp(x**3 - 2*x**2)

# Generate a range of x values
x_values = np.linspace(-2, 3, 400)  # Adjust the range and number of points as needed

# Compute the corresponding y values
y_values = [f(x) for x in x_values]

# Plot the graph
plt.plot(x_values, y_values)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Graph of f(x) = exp(x^3 - 2*x^2)')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definiere die Funktion f und ihre Ableitung f'
def f(x):
    return np.exp(x**3 - 2 * x**2)

def f_prime(x):
    return np.exp(x**3 - 2 * x**2) * (3 * x**2 - 4 * x)

# Erstelle den Wertebereich für x
x = np.linspace(-1, 2, 500)

# Berechne die Funktionswerte
y = f(x)
y_prime = f_prime(x)

# Erstelle das Plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, label="f(x) = exp(x^3 - 2x^2)")
plt.plot(x, y_prime, label="f'(x)", linestyle='--')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Plot von f(x) und f'(x)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definiere die Funktion f
def f(x1, x2):
    return -np.cos(x1**2 + x2**2 + x1 * x2)

# Erstelle ein Gitter im Bereich [-2, 2] x [-2, 2]
x1 = np.linspace(-2, 2, 500)
x2 = np.linspace(-2, 2, 500)
X1, X2 = np.meshgrid(x1, x2)

# Berechne die Funktionswerte
Z = f(X1, X2)

# Erstelle das Konturdiagramm
plt.figure(figsize=(8, 6))
contour = plt.contourf(X1, X2, Z, levels=50, cmap='viridis')
plt.colorbar(contour)
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
plt.title("Konturdiagramm von $f(x_1, x_2) = -\\cos(x_1^2 + x_2^2 + x_1 x_2)$")
plt.show()
```
