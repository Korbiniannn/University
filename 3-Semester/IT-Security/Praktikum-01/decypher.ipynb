```python
def file_to_strin(filename):
    with open(filename, 'r', encoding='utf-8') as file:
        content = file.read()

    return content
```

```python
from collections import Counter
import matplotlib.pyplot as plt

def decypher(text):

    text = file_to_strin(text)


    mapping =  {'v': 'e',
                'ü': 'n',
                'i': 's',
                'r': 'i',
                'o': 'r',
                'e': 't',
                'q': 'h',
                'ß': 'd',
                'p': 'a',
                'k': 'l',
                'm': 'u',
                'j': 'm',
                'c': 'c',
                'y': 'g',
                'g': 'o',
                's': 'f',
                'ä': 'w',
                'w': 'b',
                'ö': 'k',
                '_': 'D',
                't': 'z',
                'd': 'ü',
                'h': 'ß',
                'l': 'ä',
                'a': 'p',
                'u': 'v',
                'b': 'ö',
                'f': 'j',
                'z': 'y'}
    

    decyphered_text = ""

    for char in text:
        if char == '_':
            decyphered_text += 'D'
        elif char.lower() in mapping:
            new_char = mapping[char.lower()]
            decyphered_text += new_char.upper() if char.isupper() else new_char.lower()
        else:
            decyphered_text += char

    
    with open('decyphered_text.txt', 'w') as file:
        file.write(decyphered_text)


    return decyphered_text, mapping
```

```python
german_freq = "enisratdhulcgmoübfkwzäövpjyxqß"

text = file_to_strin('interessantesDokument.txt')

text_freq = Counter(char for char in text.lower() if char.isalpha() or char in 'äöüß_')
freq_ordered = [item[0] for item in text_freq.most_common()]
text_freq = text_freq.most_common()

characters = [item[0] for item in text_freq]
frequencies = [item[1] for item in text_freq]

# Plotting
plt.figure(figsize=(10, 5))
plt.bar(characters, frequencies, color='blue')
plt.xlabel('Characters')
plt.ylabel('Frequencies')
plt.title('Character Frequencies in Text')
plt.show()
```
