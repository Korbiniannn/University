```python
import matplotlib.pyplot as plt

# Daten
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [15, 30, 45, 10]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']

# Explode: Herausziehen des zweiten Stücks (Bananas)
explode = (0, 0.1, 0, 0)  # nur das zweite Stück wird herausgezogen

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')  # stellt sicher, dass das Pie Chart rund ist
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Erstellen von Daten
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Erstellen des Plots
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')

# Hinzufügen von Titeln und Labels
plt.title("Sinus und Cosinus")
plt.xlabel("x-Werte (Bogenmaß)")
plt.ylabel("Funktion Werte")

# Hinzufügen einer Legende
plt.legend()

# Anzeigen des Plots
plt.show()
```
