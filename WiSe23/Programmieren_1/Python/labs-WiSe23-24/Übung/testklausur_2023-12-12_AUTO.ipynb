## Quadratzahlen

Gegeben:
* Eine Liste von Zahlen.

Gesucht:
* Eine Funktion, die die Liste der Quadrate aller Zahlen aus `z` zurückliefert, die Vielfache von `3` sind.

Bitte lösen Sie mit List Comprehension.

```python
# Beispiel
in_bsp = [0, 1, 2, 3, 4]
out_bsp = [ 0, 9 ]
```

```python
# dq ... dreier quadrate
def dq(liste):
    output = []
    for x in liste:
        if x % 3 == 0:
            output.append(x**2)
    return output

dq([0, 1, 2, 3, 4])
```

```python
# assert dq(in_bsp) == out_bsp
```

## Zweitupel

Gegeben:
* eine Zahl `n`

Gesucht:
* eine Funktion, die eine Liste aller 2-Tupel der natürlichen Zahlen von `1` bis `n` zurückgibt.

Bitte lösen Sie mit List Comprehension.

```python
# Beispiel
n = 3
out_bsp = [(1, 1), (1, 2), (1, 3), (2, 1), (2, 2), (2, 3), (3, 1), (3, 2), (3, 3)]
```

```python
def zweitupel(n):
    L = []
    for x in range(1, n + 1):
        for y in range(1, n + 1):
            L.append((x, y))
    return L

zweitupel(3)
```

```python
# assert zweitupel(3) == out_bsp
```
